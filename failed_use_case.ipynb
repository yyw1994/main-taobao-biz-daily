# Take Screenshot
Define a function that uses Selenium to take a screenshot after logging in.

```python
# Take Screenshot
import os
from datetime import datetime

def take_screenshot(driver, file_path):
    """
    Function to take a screenshot using Selenium and save it to a folder with the current date as its name.
    
    Parameters:
    driver (webdriver): The Selenium WebDriver instance.
    file_path (str): The file path where the screenshot will be saved.
    """
    # Get the current date
    current_date = datetime.now().strftime('%Y-%m-%d')
    
    # Create a folder with the current date as its name if it doesn't exist
    if not os.path.exists(current_date):
        os.makedirs(current_date)
    
    # Construct the full file path
    full_file_path = os.path.join(current_date, file_path)
    
    # Take a screenshot and save it to the specified file path
    driver.save_screenshot(full_file_path)
```

```python
import os
import time
from datetime import datetime
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from apscheduler.schedulers.blocking import BlockingScheduler

# Function to log in to Jisilu and take a screenshot
def login_and_screenshot(email, password):
    # Set up the Chrome driver
    options = uc.ChromeOptions()
    options.add_argument("--headless")  # Run in headless mode
    driver = uc.Chrome(options=options)
    
    try:
        # Open Jisilu login page
        driver.get("https://www.jisilu.cn/account/login/")
        
        # Wait for the email input field to be present
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.ID, "aw-login-user-name"))
        )
        
        # Find the email input field and enter the email
        email_field = driver.find_element(By.ID, "aw-login-user-name")
        email_field.send_keys(email)
        
        # Find the password input field and enter the password
        password_field = driver.find_element(By.ID, "aw-login-user-password")
        password_field.send_keys(password)
        
        # Find the login button and click it
        login_button = driver.find_element(By.XPATH, "//a[@class='btn btn-large btn-primary']")
        login_button.click()
        
        # Wait for the login process to complete
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, "//a[@href='/account/logout/']"))
        )
        
        # Take a screenshot
        take_screenshot(driver, "screenshot.png")
        
    finally:
        driver.quit()

# Function to take a screenshot and save it in a folder with the current date
def take_screenshot(driver, file_name):
    current_date = datetime.now().strftime('%Y-%m-%d')
    if not os.path.exists(current_date):
        os.makedirs(current_date)
    file_path = os.path.join(current_date, file_name)
    driver.save_screenshot(file_path)

# Function to schedule the task
def schedule_task(email, password):
    scheduler = BlockingScheduler()
    scheduler.add_job(login_and_screenshot, 'cron', hour=1, args=[email, password])
    scheduler.start()

# Replace with your Jisilu account email and password
email = "yyw1994"
password = "St7950819"

# Schedule the task
# schedule_task(email, password)

login_and_screenshot(email, password)
```
